```python
import requests
```

```python
link="https://www.fifaratings.com/country/india"
```

```python
response=requests.get(link)
```

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
name=soup.select('.entry-font-narrow .text-dark')
```

```python
name
```

```python
club=soup.select('.text-dark+ .text-dark')
```

```python
club
```

```python
potential=soup.select('td:nth-child(4) .attribute-box')
```

```python
potential
```

```python
position=soup.select('br+ .text-dark')
```

```python
position
```

```python
import pandas
```

```python
football_players=pandas.DataFrame(name,columns=["Name"])
```

```python
football_players
```

```python
football_players["club"]=club
```

```python
football_players["Position"]=position
```

```python
football_players["Potential"]=potential
```
